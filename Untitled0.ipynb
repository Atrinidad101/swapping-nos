```python
import math

def calculate_distance(angle, velocity):
    angle_radians = math.radians(angle)
    distance = (velocity ** 2 * math.sin(2 * angle_radians)) / 9.8
    return distance

def shooting_game():
    print("Welcome to the Shooting Game!")
    target_distance = 100  # Set the target distance
    print(f"Target is {target_distance} meters away.")

    while True:
        try:
            angle = float(input("Enter the angle (in degrees): "))
            velocity = float(input("Enter the velocity (in meters/second): "))
            if angle < 0 or velocity < 0:
                raise ValueError
            break
        except ValueError:
            print("Please enter valid positive numerical values for angle and velocity.")

    distance = calculate_distance(angle, velocity)
    if abs(distance - target_distance) <= 5:
        print("Congratulations! You hit the target!")
    else:
        print(f"You missed! The projectile landed {distance:.2f} meters away from the target.")

shooting_game()
```
